# Threads VS Process

## Threads ex.

```python
def do_nothing():
    while True:
        pass
```

```python
for i in range(20):
    th.Thread(target=do_nothing).start()
    
th.enumerate()
```

### Import libs, manage process (cores)

```python
import multiprocessing as mp # Manage cores
```

```python
mp.cpu_count()
```

```python
mp.current_process()
```

## Process Ex.

```python
for i in range(20):
    mp.Process(target=do_nothing).start()
    
```
